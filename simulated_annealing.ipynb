```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def distance(N,route,x,y):
  dis = 0.0
  for i in range(N-1):
    dx = x[route[i]]-x[route[i+1]]
    dy = y[route[i]]-y[route[i+1]]
    dis = dis + np.sqrt(np.power(dx,2)+np.power(dy,2))
  dx = x[route[N-1]]-x[route[0]]
  dy = y[route[N-1]]-y[route[0]]
  dis = dis + np.sqrt(np.power(dx,2)+np.power(dy,2))
  return dis
```

```python
N = 200 # number of cities
random.seed()
# initialization
x = np.zeros(N)
y = np.zeros(N)
x[0] = 0.50
y[0] = 0.50
for i in range(1,N-1):
  x[i] = random.random()
  y[i] = random.random()
# print the cities
plt.scatter(x[0],y[0])
plt.scatter(x[1:],y[1:])
plt.show()
```

```python
k = 0
route = np.arange(N)
dis = distance(N,route,x,y)
print(k,dis)
diso = dis
T = 100.0
dt = 0.999
min_dis = dis
epoch = []
hist = []
while T>0.0001:
  k = k+1
  city1 = random.randint(1,N-1)
  city2 = random.randint(1,N-1)
  while city1==city2:
    city2 = random.randint(1,N-1)
  r1 = route[city1]
  r2 = route[city2]
  route[city1] = r2
  route[city2] = r1
  dis = distance(N,route,x,y)
  prob = np.exp(-(dis-diso)/T)
  if prob>=random.random():
    diso = dis
  else:
    route[city1] = r1
    route[city2] = r2
  if dis<min_dis:
    min_dis = dis
  print(k,T,diso,min_dis)
  T = T*dt
  epoch.append(k)
  hist.append(diso)
```

```python
plt.plot(epoch,hist)
plt.show()
```
