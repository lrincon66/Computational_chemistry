```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
# Ackley Function
def fun(x):
  return -20.0*np.exp(-0.2*np.sqrt(0.5*(np.power(x[0],2)+np.power(x[1],2))))-np.exp(0.5*(np.cos(2.0*np.pi*x[0])+np.cos(2.0*np.pi*x[1])))+np.exp(1.0)+20.0
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = fun([X,Y])
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False)
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
fig, ax = plt.subplots()
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = fun([X,Y])
im = ax.imshow(Z, interpolation='bilinear', origin='lower', extent=(-5, 5, -5, 5))
# Plot contour
CS = ax.contour(X,Y,Z)
# Add a color bar which maps values to colors.
CBI = fig.colorbar(im, shrink=0.8)
plt.show()
```

```python
def grad(x):
  gradx = np.exp(-0.2*np.sqrt(0.5*(np.power(x[0],2)+np.power(x[1],2))))*(x[0]/np.sqrt(0.5*(np.power(x[0],2)+np.power(x[1],2)))) + np.exp(0.5*(np.cos(2.0*np.pi*x[0])+np.cos(2.0*np.pi*x[1])))*np.pi*np.sin(2.0*np.pi*x[0])
  grady = np.exp(-0.2*np.sqrt(0.5*(np.power(x[0],2)+np.power(x[1],2))))*(x[1]/np.sqrt(0.5*(np.power(x[0],2)+np.power(x[1],2)))) + np.exp(0.5*(np.cos(2.0*np.pi*x[0])+np.cos(2.0*np.pi*x[1])))*np.pi*np.sin(2.0*np.pi*x[1])
  return([gradx,grady])
```

```python
x0 = [2.0,2.0]
res = minimize(fun, x0, method="L-BFGS-B",jac=grad)
print(res.x,res.fun)
```
